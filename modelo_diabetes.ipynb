```python
# Importação das bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leitura do arquivo CSV
df = pd.read_csv('data/diabetes.csv')
```

```python
# Visualização da estrutura do DataFrame
df.info()
```

```python
# Amostra das 10 primeiras linhas do DataFrame
df.head(10)
```

```python
df['genero'].unique()
```

```python
# Remoção da coluna id_paciente
df.drop(columns='id_paciente', axis=1, inplace=True)
```

```python
# Conversão de uma variável categórica (genero) em numérica usando One-Hot Encoding
df = pd.get_dummies(df, columns=['genero'], dtype='int64')
```

```python
df.head()
```

```python
# Apresentação do mapa de calor da correlação entre as variáveis
sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Mapa de calor só com a variável target
sns.heatmap(df.corr()[['resultado']].sort_values(by='resultado', ascending=False), vmin=-1, vmax=1, annot=True, cmap='coolwarm')
```

```python
# Plotagem do scatterplot com distribuição
pd.plotting.scatter_matrix(df, alpha=0.2, figsize=(12, 10), diagonal='kde')
plt.show()
```

```python
# Histograma de todas as variáveis
df.hist(layout=(2, 3), figsize=(12, 8))
plt.show()
```
